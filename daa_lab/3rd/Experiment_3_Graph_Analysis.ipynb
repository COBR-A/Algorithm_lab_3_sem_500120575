```python
import matplotlib.pyplot as plt
```

```python
n=[64,128,256,512,1024,2048]
Traditional=[0.001,0.008,0.051,0.299,3.9,57.309]
Strassen=[0.098,0.347,2.191,14.664,112.095,1025.25]
```

```python
plt.plot(n, Traditional, marker='o', label='Traditional Matrix Multiplication')
plt.plot(n, Strassen, marker='o', label='Strassen\'s Matrix Multiplication')


plt.xlabel('Matrix Size (n*n)')
plt.ylabel('Time (seconds)')
plt.title('Algorithm Performance Comparison')
plt.legend()
plt.grid(True)

plt.show()
```
